Numpy 배열에서 정보를 얻는 기본 속성은 다음과 같습니다.

- **ndim** : 차원(축)의 수
- **shape** : 각 차원의 크기(튜플)
- **size** : 전체 요소의 개수, shape의 각 항목의 곱
- **dtype** : 요소의 데이터 타입
- **itemsize** : 각 요소의 바이트 크기

이미지는 여러 개의 픽셀들의 값으로 구성되기 때문에 수많은 행과 열로 구성된 픽셀 데이터의 모음으로 볼 수 있습니다. 

어떤 이미지의 타입을 살펴보면 다음과 같습니다.

```python
import cv2
print(cv2.__version__)


import os

# 현재 작업 디렉토리 확인
current_dir = os.getcwd()

# 이미지 파일의 상대 경로
image_path = 'img\\mika.jpg'

# 이미지 파일의 절대 경로 생성
absolute_path = os.path.join(current_dir, image_path)

# 이미지 읽기
img = cv2.imread(absolute_path)
#img = cv2.imread('../img/mika.jpg')
type(img)
```

ndarray는 N-Dimensional Array의 약자로 다차원 배열을 의미합니다. OpenCV는 기본적으로 이미지를 3차원 배열인 '행(row)x열(column)x채널(channel)'로 표현합니다.

```python
print('이미지 차원 : ',img.ndim)
print('이미지 크기 : ',img.shape)
print('이미지 전체 요소 개수 : ',img.size) # 전체 요소의 개수
print('이미지 데이터 타입 : ',img.dtype) # 부호없는 8비트 ( 이미지 픽셀은 음수나 소수점이 없다)
print('이미지 각 요소의 바이트 크기 : ',img.itemsize) # 각 요소의 크기가 1바이트
```

#### 1-2. Numpy 배열 생성

Numpy 배열은 값을 가지고 생성하는 방법, 크기만 지정해서 생성하는 방법으로 나뉩니다. 크기만 지정해서 생성하는 방법은 다시 특정한 초기 값을 모든 요소에 지정하는 경우와 값의 범위를 지정해서 순차적으로 증가 또는 감소하는 값을 갖게하는 방법으로 나눌 수 있습니다.

- 값으로 생성 : array()
- 초기 값으로 생성 : empty(), zeros(), ones(), full()
- 기존 배열로 생성 : empty_like(), zeros_like(), ones_like(), full_like()
- 순차적인 값으로 생성 : arange()
- 난수로 생성 : random.rand(), random.randn()

- **값으로 생성**

배열 생성에 사용할 값을 가지고 있는 경우에는 numpy.array() 함수로 쉽게 생성할 수 있습니다.

- numpy.array(list \[, dtype\]) : 지정한 값들로 Numpy 배열을 생성
    - list : 배열 생성에 사용할 값을 갖는 파이썬 리스트 객체
    - dtype : 데이터 타입 ( 생략하면 값에 의해 자동 결정 )
        - int8, int16, int32, int64 : 부호 있는 정수
        - uint8, uint16, uint32, uint64 : 부호 없는 정수
        - float16, float32, float64, float128 : 부동 소수점을 갖는 실수
        - complex64, complex128, complex256 : 부동 소수점을 갖는 복소수
        - bool : 불리언

이 외에도 훨씬 더 많은 dtype이 존재합니다. 

```python
import numpy as np
a = np.array([1,2,3,4])
print('a :',a)
print('a.dtype : ',a.dtype)
print('a.shape : ',a.shape)
```

```python
c = np.array([1,2,3.14,4])
print(c)
print(c.dtype)
```

소수가 함께 있는 array에는 dtype이 float형태로 나옵니다.<br>
명시적으로 dtype를 지정하고 싶은 경우에는 다음과 같이 지정하고픈 dtype을 옵션으로 주면 됩니다.

```python
d = np.array([1,2,3,4], dtype=np.float32)
print('d : ',d)
print('d.dtype : ',d.dtype)
```

- **크기와 초기 값으로 생성**

Numpy 배열을 생성할 때 사용할 값을 가지고 있지 않은 경우가 많기 때문에 초기 값을 지정해서 생성하는 방법을 사용합니다.

- numpy.empty(shape \[, dtype\]) : 초기화되지 않는 값(쓰레기 값)으로 배열 생성
    - shape : 튜플, 배열의 각 차수의 크기 지정
- numpy.zeros(shape \[,dtype\]) : 0으로 초기화된 배열 생성
- numpy.ones(shape \[,dtype\]) : 1로 초기화된 배열 생성
- numpy.full(shape, fill_value \[,dtype\]) : fill_value로 초기화된 배열 생성

```python
b = np.zeros((2,3))
print(b)
print(b.dtype)
```

```python
# dtype을 지정해주면 지정해준 type을 따라갑니다.
c = np.zeros((2,3), dtype = np.int8)
print(c.dtype)
```

```python
d = np.ones((2,3), dtype = np.int16)
print(d)
print(d.dtype)
```

새로운 배열을 생성할 때 기존에 있던 배열과 같은 크기의 배열을 만들때에는 위의 함수들에 \_like를 붙인 함수를 사용합니다.

empty_like, zeros_like, ones_like, full_like

```python
import numpy as np
# 예제 배열 생성
original_array = np.array([[1, 2, 3], [4, 5, 6]])

# empty_like: 주어진 배열과 동일한 shape 및 dtype을 가진 빈 배열 생성 -> 값이 들어있는데 이상한값들이네, empty가아님
empty_array = np.empty_like(original_array)
print("Empty Array:")
print(empty_array)

# zeros_like: 주어진 배열과 동일한 shape 및 dtype을 가진 0으로 초기화된 배열 생성
zeros_array = np.zeros_like(original_array)
print("\nZeros Array:")
print(zeros_array)
```

- **시퀀스와 난수로 생성**

Numpy 배열을 생성하는 방법 중에는 일정한 범위 내에서 순차!!적인 값을 갖게하는 방법과 난수!!로 채우는 방법이 있습니다.

- numpy.arange(\[start=0, \] stop \[, step=1, dtype = float64\]) : 순차적인 값으로 생성
    - start : 시작 값
    - stop : 종료 값, 범위에 포함되는 수는 ~ stop -1
    - step : 증가 값
    - dtype : 데이터 타입
- numpy.random.rand(array크기) : 0과 1 사이의 무작위 수 생성
    - array크기를 생략하면 난수 1개 반환, 나머지는 해당 크기만큼 값을 반환
- numpy.random.randn(array크기) : 표준 정규 분포를 따르는 무작위 수 생성

```python
import numpy as np
a = np.arange(5)
print(a)
print(a.dtype)
print(a.shape)
```

```python
import numpy as np
c = np.arange(3,9,2)
print(c)
```

마지막 9는 범위에 포함되지 않습니다. -1인 8의 값까지에서 종료되기 때문입니다.

```python
print(np.random.rand())
print(np.random.randn())
print(np.random.rand(2,3))
print(np.random.randn(2,3))

'''
이 함수들은 결과 값이 소수점을 갖고 특정 범위 내에서 난수를 추출하기 때문에
이미지 작업에 필요한 원하는 범위 내에서 난수를 발생하기 위해서는
브로드캐스팅 연산과 dtype 변경이 필요할 때가 많습니다.
'''
```

#### 1-4. 차원 변경

- ndarray.reshape(newshape) : ndarray의 shape를 newshape로 차원 변경
- numpy.reshape(ndarray, newshape) : ndarray의 shape를 newshape로 차원 변경
    - ndarray : 원본 배열 객체
    - newshape : 변경하고자 하는 새로운 shape
- numpy.ravel(ndarray) : 1차원 배열로 차원 변경
    - ndarray :변경할 원본 배열
- ndarray.T : 전치배열 (transpose)    

arange함수는 1차원만 생성 가능합니다.

```python
a = np.arange(6)
print('a : ',a)
```

이를 2차원으로 바꾸어보겠습니다.

```python
b = a.reshape(2,3)
print('b : ',b)
c = np.reshape(a,(2,3))
print('c : ',c)
```

다음은 reshape와 ravel을 이용하여 예시 데이터를 재정렬해보겠습니다.

```python
# 1차원으로 재정렬 : reshape, ravel
f = np.zeros((2,3))
print(f)
print("------")
print(f.reshape((6,)))
print("------")
print(f.reshape(-1))
print("------")
print(np.ravel(f)) # 1차원 배열로 변경
```

전치배열을 이용하여 행과 열의 shape를 바꾸어보겠습니다.

```python
g = np.arange(10).reshape(2,-1)
print(g)
print(g.T)
```

#### 1-6. 인덱싱과 슬라이싱

Numpy 배열은 파이썬의 리스트처럼 인덱스로 각 요소에 접근 할 수 있습니다. 배열의 차원에 따라서 인덱스의 개수도 달라집니다.

- 1차원

```python
a = np.arange(10)
print(a)
print(a[5])
print(a[2:5]) # 3번재부터 5번째까지 출력
print(a[5:]) # 6번째부터 끝까지 출력
```

- 2차원

```python
b = np.arange(12).reshape(3,4)
print(b)
print(b[1]) # b의 두번째 행
print(b[1,2]) # b의 2번째 행의 3번째 요소
```

슬라이싱으로 전체 배열 중 일부를 다른 변수에 할당하는 경우 값의 변경은 원본에도 그대로 반영됩니다. 만약 바뀌는걸 원치 않는다면 **ndarray.copy()함수**를 사용해여 복제본을 만들어두면 됩니다.

```python
print(b)
```

```python
bb = b[0:2, 1:3] # b에서 슬라이싱해온 값을 bb변수에 저장
print(bb)
```

```python
bb[0] = 99  # bb 변수의 0행의 값을 99로 변경
print(b)
```

이렇게 바뀌기 때문에 복제본을 이용하면 원본의 값을 변화시키지 않고 원하는 값을 얻을 수 있습니다.

#### 1-7. 팬시 인덱싱

배열 인덱스에 다른 배열을 전달해서 원하는 요소를 선택하는 방법을 **팬시 인덱싱**이라고 합니다. 전달하는 배열에 숫자를 포함하고 있으면 해당 인덱스에 맞게 선택되고, 배열에 boolean 값을 포함하면 True 인 값을 갖는 요소만 선택됩니다.

```python
a = np.arange(5)
print(a)
print(a[[1,3]])
print(a[[True, False, True, False, True]]) # True값만 출력
```

Numpy 배열에 비교 연산을 하면 개별 요소들이 조건을 만족하는지를 알 수 있습니다. 반대로 boolean 값을 갖는 배열을 배열의 인덱스 대신 사용하면 True 값 위치의 값들만 얻을 수 있는데, 이 둘을 한번에 합해서 실행하면 원하는 조건의 값만을 얻을 수 있습니다. 

```python
import numpy as np

a = np.arange(10)
print('a : ',a)
b = a > 5
print('b : ',b)
```

다음과 같이 조건을 지정한 후 먼저 a[b]값을 구해보면 다음과 같습니다.

```python
print('a[b] : ',a[b])
```

b에서 True인 값의 숫자만 출력된 것을 확인할 수 있습니다.

다음은 b 조건에 따른 a의 값을 변경시켜보겠습니다. a 중에 5 초과인 원소들을 1로 바꾸어봅시다.

```python
a[a>5] = 1
print('변경된 a :',a)
```

#### 1-8. 병합과 분리

2개 이상의 Numpy 배열을 병합하는 방법은 크게 단순히 배열을 이어 붙여서 크기를 키우는 방법과 새로운 차원을 만들어 서로서로 끼워넣는 방법, 2가지가 있습니다. 이를 이해하기 위해서는 축(axis)의 개념을 알아야 하는데 shape가 (10, 20, 3)이면 3개의 축이고 10은 0번축, 20은 1번축으로 생각하고, **축을 기준으로 작업을 한다는 뜻은 shape의 각 순서에 따라 작업을 한다는 뜻**입니다. 

- numpy.hstack(arrays) : arrays 배열을 수평으로 병합
- numpy.vstack(arrays) : arrays 배열을 수직으로 병합
- numpy.concatenate(arrays, axis=0) : arrays 배열을 지정한 축 기준으로 병합
- numpy.stack(arrays, axis=0) : arrays 배열을 새로운 축으로 병합
    - arrays : 병합 대상 배열(튜플)
    - axis : 작업할 대상 축 번호

- **vstack, hstack**

예시 데이터를 먼저 생성 후,

```python
a = np.arange(4).reshape(2,2)
print(a)
b = np.arange(10, 14).reshape(2,2)
print(b)
```

이를 병합해봅시다.

- 수직 병합

```python
print(np.vstack((a,b))) # vertical
```

```python
print(np.concatenate((a,b),0))
```

- 수평 병합

```python
print(np.hstack((a,b))) # horizontal
```

```python
print(np.concatenate((a,b),1)) # 수평으로 병합
```

- numpy.split(array, indice, axis = 0 ) : array 배열을 axis 축으로 분리
    - array : 분리할 배열
    - indice : 분리할 개수 또는 인덱스 ( 어떻게 나눌지를 정하는 인자로 정수 = 배열을 그 수로 나눔, 1차워 배열 = 나눌 인덱스 )
    - axis : 기준 축 번호

```python
b = np.arange(12).reshape(4,3)
print(b)
```

```python
print(np.vsplit(b,2)) # 수직으로 나눔
```

```python
print(np.split(b, 2, 0)) # 0번 축을 기준으로 2개로 나눔
```

#### 2-4. subplot

각각의 그래프를 분리해서 따로 그려야 할 때 사용합니다. 이 subplot을 이용하여 다양한 그래프를 한 번에 나타낼 수 있습니다.

```python
x = np.arange(10)

plt.subplot(2,2,1)  #2행 2열 중에 1번째
plt.plot(x,x**2)

plt.subplot(2,2,2)  #2행 2열 중에 2번째
plt.plot(x,x*5)

plt.subplot(223)    #2행 2열 중에 3번째
plt.plot(x, np.sin(x))

plt.subplot(224)    #2행 2열 중에 4번째
plt.plot(x,np.cos(x))

plt.show()
```

#### 2-5. 이미지 표시

plt.plot() 대신에 **plt.imshow() 함수**를 호출하면 OpenCV로 읽어들인 이미지를 그래프 영역에 출력할 수 있습니다. 이는 데이콘에서 파일 올리기를 할 때도 매우 중요한 기능입니다! cv2.imshow()함수를 이용하면 데이콘 화면에서는 안보이더라구요 ㅠㅠ..

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('../img/woman.jpg')

plt.imshow(img) # 이미지 표시
plt.show()
```

하지만 그냥 출력해버리면 다음과 같이 이미지의 색이 다른 것을 확인할 수 있습니다. **plt.imshow() 함수는 컬러 이미지를 RGB 순으로 해석하지만 OpenCV는 BGR순으로 인식하기 때문입니다. 따라서 컬러를 뒤집어서 나타내주어야 합니다.**

```python
img = cv2.imread('../img/woman.jpg')


plt.imshow(img[:,:,::-1]) # 이미지 컬러 채널을 뒤집어서 표시 ( 3번째가 채널 )
# plt.imshow(img[:,:,(2,1,0)])  위와 같은 결과
plt.xticks([])
plt.yticks([])
plt.show()
```

이미지들을 앞서 배웠던 subplot을 이용하여 다시 그려보겠습니다.

```python
img1 = cv2.imread('../img/woman2.jpg')
img2 = cv2.imread('../img/woman3.jpg')
img3 = cv2.imread('../img/woman4.jpg')


plt.subplot(1,3,1)  #1행 3열 중에 1번째
plt.imshow(img1[:,:,(2,1,0)])
plt.xticks([]); plt.yticks([])

plt.subplot(1,3,2)  #1행 3열 중에 2번째
plt.imshow(img2[:,:,(2,1,0)])
plt.xticks([]); plt.yticks([])

plt.subplot(1,3,3)    #1행 3열 중에 3번째
plt.imshow(img3[:,:,(2,1,0)])
plt.xticks([]); plt.yticks([])

plt.show()
```
